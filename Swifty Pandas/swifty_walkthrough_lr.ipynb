## Why use Pandas?

Excel: It's Me. <br>
Pandas: Hi. <br>
Excel: I'm the problem, it's me. <br>

#### Scalability

Pandas: Can handle large datasets more efficiently than Excel, which can become slow or crash with very large files.<br>
Excel: Limited to handling smaller datasets due to memory and performance constraints.

#### Automation

Pandas: Ideal for automating repetitive tasks with scripts, reducing manual effort and minimizing errors.<br>
Excel: Requires manual effort for many tasks, which can be time-consuming and prone to human error.

#### Reproducibility

Pandas: Allows you to create reproducible pipelines with code that can be shared and reused.<br>
Excel: Changes can be made manually without a clear record, making it harder to reproduce results.

####  Version Control

Pandas: Code can be version controlled using tools like Git, making it easier to track changes and collaborate with others.<br>
Excel: Changes are tracked within the file, which can be difficult to manage and collaborate on in a controlled manner.

####  Repeatability and Documentation

Pandas: Code serves as documentation, showing exactly how data was transformed and analyzed.<br>
Excel: Transformations and analyses are less transparent, often embedded in cells without clear documentation.

####  Flexibility

Pandas: Can handle various data formats (e.g., CSV, Excel, SQL databases, JSON) and supports complex data structures.<br>
Excel: Primarily focused on tabular data and limited in handling different data formats and structures.

## Dataset
This dataset contains data from Spotify's API for Taylor Swift songs/albums.

<b>name</b>  - the name of the song <br>
<b>album</b>  - the name of the album <br>
<b>release_date</b>  - the day month and year the album was released<br>
<b>track number</b>  - the order the song appears on the album<br>
<b>id</b>  - the Spotify id for the song<br>
<b>uri</b>  - the Spotify uri for the song<br>
<b>acousticness</b>  - A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.<br>
<b>danceability</b>  - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.<br>
<b>energy</b>  - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. <br>
<b>instrumentalness</b>  - Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. <br>
<b>liveness</b>  - Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.<br>
<b>loudness</b>  - The overall loudness of a track in decibels (dB). <br>
<b>speechiness</b>  - Detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value.<br>
<b>tempo</b>  - The overall estimated tempo of a track in beats per minute (BPM).<br>
<b>valence</b>  - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).<br>
<b>popularity</b>  - The popularity of the song from 0 to 100<br>
<b>duration_ms</b> - The duration of the track in milliseconds.<br>

### Exercise 2: Data Cleaning

```python
# Keep the dataset in view so we can remember the columns
df.head(1)
```

```python
# Convert release date to a nice date format, and create a more intuitive duration column
df['release_date'] = pd.to_datetime(df['release_date'], format='mixed', yearfirst=True)
df['duration_minutes'] = df['duration_ms'] / 60000
```

```python
# Sort the dataframe by release date and track number, see the index (1st column, no header) isn't in order now, so we'll reset that further down
df = df.sort_values(by=['release_date','track_number'], ascending=True)
df.head()
```

```python
# Drop the spotify ID, URI and duration_ms, don't think we'll need them for this analysis, and reset the index
df = df.drop(['id','uri','duration_ms'],axis=1)
df = df.reset_index(drop=True)
df.head()
```

### Exercise 3: Pandas Functions

```python
# Group by 'album' and calculate the mean energy for each album
df.groupby('album')['energy'].mean().reset_index().sort_values(by='energy', ascending=False).reset_index(drop=True)
```

### Exercise 4: EDA

```python
# Create a list of numerical columns to use later
numerical_columns = df.select_dtypes(include='number').columns.tolist()

# Remove 'track_number' from the list, it's not really useful information for visuals
numerical_columns.remove('track_number')
numerical_columns
```

#### Variance

Low variance columns in datasets are often not useful for models because they don't provide sufficient information or variability to help the model distinguish between different samples.

<b>Constant Values</b>

For example, a column that has the same value for every data point (e.g., all zeros or all ones) provides no discriminatory information.

<b>Redundant Information

Noise

```python
# Take a look at each column's variance
variances = df[numerical_columns].var().sort_values()
variances
```

```python
# Set a variance threshold, and only select columns that are above that
threshold = 0.01
numerical_columns = variances[variances > threshold].index.tolist()
numerical_columns
```

#### Correlations

Generally, highly correlated features are bad for certain models:

<br>
<b>Difficulty in Understanding the Relationship</b> 
<br>
<br>
When variables are highly correlated, it obscures the effect of individual variables because their effects are intertwined, making it difficult to determine which predictors are actually driving changes in the dependent variable.
<br>
<br>
<b>Redundant Information</b>
<br>
<br>
High multicollinearity indicates that some variables provide redundant information - they don't add value to the model and can complicate the interpretation without improving the model's predictive power.

<b> Inflated Standard Errors, Unstable Coefficient Estimates etc.

```python
# Create a correlation matrix of all the numerical features
correlation_matrix = df[numerical_columns].corr()

# Plot it
plt.figure(figsize=(10, 8)) # controls the size of the plot
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f', linewidths=0.5) # plot the heatmap from the Seaborn package
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# Darker squares are more positively correlated, lighter squares are more negatively correlated
# Acousticness negatively correlated with energy, which makes sense - as acoutsticness goes up, energy goes down
# Loudness is positively correlated with energy - as loudness goes up, so does energy
```

```python
# We can see the above correlations as scatterplots below - takes about 25 seconds to run, doing a lot under the hood
sns.pairplot(df[numerical_columns], diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()

# Danceability and Energy has a small to medium positive correlation with Valence (positivity)
# Obvious positive relationship between Energy and Loudness
# Negative relationship between Energy and Acoustic ness - As Acousticness decreases, energy increase
# On the Diagonal you have a smoothed histogram which just represents the distribution of that feature
```

```python
# Plot energy across albums - as we've ordered the dataset by release date, it's sort of like a time series
plt.figure(figsize=(8, 8))
sns.boxplot(data=df, x='album', y='energy', palette="crest", hue='album')
plt.title('Energy Across Albums')
plt.ylabel('Energy')
plt.xticks(rotation=90)
plt.show()

# Reading a boxplot - the box represents the middle 50% of the data, the line in the box in the median, then you have 25% either side - smaller boxes represent less variance in the data
# Energy of the albums was pretty consistent until the Lover Era, then it dropped considerably for Folklore +
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='album', y='valence', palette='crest', hue='album')
plt.title('Positivity Across Albums')
plt.ylabel('Positivity')
plt.xticks(rotation=90)
plt.show()

# A definite shift over time towards more negativity in the songs - midnights and TTPS
```

```python
# Violin plot to show distribution of Popularity across different albums - same idea, different shape
plt.figure(figsize=(10, 8))
sns.violinplot(data=df, x='album', y='popularity', palette='crest', density_norm='width', hue='album')
plt.title('Popularity Distribution Across Albums')
plt.xlabel('Album')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.show()

# Lover had some absolute bangers on it, generally on an upward tick now 
```

### Exercise 5: Predicting an album's popularity

```python
# Import packages for machine learning
from sklearn.linear_model import LinearRegression # Simple model, probably beyond the scope of the session to learn how the model works in depth 
from sklearn.preprocessing import StandardScaler
```

### Example Linear Regression (Using TV Revenue to predict Sales)

<img src="regression_example.png" alt="Example Image" width="700px"/>

\begin{align*}
\text{Sales} &= \text{intercept term} (\beta_0)\\
&+ \beta_1 \cdot \text{TV}
\end{align*}

Mathematically, the intercept represents the point where the regression line crosses the y-axis (zero)

The $\beta$ refer to coefficients (floats) that the model learns, these get multiplied by the value of the feature.

### Predicting The Tortured Poets Department's Popularity

\begin{align*}
\text{Predicted Popularity} &=  \beta_0 \\
&+ \beta_1 \cdot \text{danceability} \\
&+ \beta_2 \cdot \text{energy} \\
&+ \beta_3 \cdot \text{valence} \\
&+ \beta_4 \cdot \text{acousticness} \\
&+ \beta_5 \cdot \text{duration\_minutes} \\
&+ \beta_6 \cdot \text{loudness} \\
&+ \beta_7 \cdot \text{tempo}
\end{align*}

```python
# Transform the data, take the mean for each feature, by album, assign it to the grouped_df variable
grouped_df = df.groupby('album')[numerical_columns].mean().reset_index()
grouped_df
```

```python
# Initialize the scaler - it's a good idea to transform the features so that they are all on the same scale when doing a Linear Regression
# Average Tempo is 120, average daceability is 0.58 - huge difference and the model will be impacted if they aren't on the same scale and it makes the feature importance harder to interpret
scaler = StandardScaler()

# Features that will be input to the model and the target (what we are trying to predict)
features = ['danceability', 'energy', 'valence', 'acousticness', 'duration_minutes', 'loudness', 'tempo']
target = 'popularity'

# Scale the features, which gives everything a mean of zero and a standard deviation of 1 (approx) - everything gets centered around zero
grouped_df[features] = scaler.fit_transform(grouped_df[features])
grouped_df
```

```python
# We'll train the model on every album except TTPD, and then try and predict TTPD's popularity
train_df = grouped_df[grouped_df['album'] != 'THE TORTURED POETS DEPARTMENT']
train_df
```

```python
test_df = grouped_df[grouped_df['album'] == 'THE TORTURED POETS DEPARTMENT']
test_df
```

```python
# Now we need to split the data into features and target. X usually denotes the feature set
X_train = train_df[features]
X_train
```

```python
# y usually denotes the target variable, so we'll call this y_train. This is the popularity of all the albums in the training set. 
y_train = train_df[target]
y_train
```

```python
# X_test are the feature values for TTPD, the model doesn't see this data, the model only sees the training data, we'll use this data after we've trained the model to see how it performs
X_test = test_df[features]
X_test
```

```python
# y_test is the actually popularity of TTPD
y_test = test_df[target]
y_test
```

```python
# Initialize and train the model. Train the model on X_train and y_train - The model learns the relationships between the input (X) and the output (y)
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions. For the information we have about TTPD (X_test), make a prediction on it's popularity, save it to y_pred (the predicted value for y)
y_pred = model.predict(X_test)

# Store the results in a dataframe so we can compare how good our prediction was
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
results
```

Underfitting the problem slightly, so we probably need more data/features/model complexity to improve predictions. There is also a time-component to the model - her popularity seems to increase per album, and we have nothing in the model for that.

### Exercise 6: Model Parameters/Feature Importance

```python
# Pull out the model coefficents so we can see which features have the largest impact on the model prediction
coef_table = pd.DataFrame(list(X_train.columns), columns=["feature"])
coef_table.insert(len(coef_table.columns),"coefficient",model.coef_.transpose())
coef_table
```

The further away from zero, the more of an impact on the target (popularity) a feature has.
